```python
import os
import wave
import numpy as np
import matplotlib.pyplot as plt
```

```python
CutTimeDef = 20
path = r"E:\MusicEmotionRec\WavData\Calm to Released"
files = os.listdir(path)
files = [path + "\\" + f for f in files if f.endswith('.wav')]

for i in range(len(files)):
    FileName = files[i]
    print("CutFile File Name is ",FileName)
    f = wave.open(r""+FileName, "rb")
    params = f.getparams()
    nchannels,sampwidth,framerate,nframes = params[:4]
    CutFrameNum = framerate * CutTimeDef
        
    print("CutFrameNum=%d" %(CutFrameNum))#20s的采样频率
    print("nchannels=%d" %(nchannels))#声道数
    print("sampwidth=%d" % (sampwidth))#量化位数
    print("framerate=%d" % (framerate))#采样频率
    print("nframes=%d" % (nframes))#采样点数
    str_data = f.readframes(nframes)
    f.close()
        
    wave_data = np.fromstring(str_data, dtype=np.short)#音频波形数据转化为数组
    #将数组改为两列，行数自动匹配
    wave_data.shape = -1, 2
    #转置
    wave_data = wave_data.T
    temp_data = wave_data.T 
    StepNum = CutFrameNum
    StepTotalNum = 0
    haha = 0
    while StepTotalNum < nframes:
        print("Stemp=%d" % (haha))
        FileName = "E:\\MusicEmotionRec\\WavData\\CalmtoReleasedPeace\\" + files[i][43:] +"-"+ str(haha+1) + ".wav"
        print(FileName)
        temp_dataTemp = temp_data[StepNum * (haha):StepNum * (haha + 1)]
        haha = haha + 1
        StepTotalNum = haha * StepNum
        temp_dataTemp.shape = 1, -1
        temp_dataTemp = temp_dataTemp.astype(np.short)
        f = wave.open(FileName, 'wb')
        f.setnchannels(nchannels)
        f.setsampwidth(sampwidth)
        f.setframerate(framerate)
        f.writeframes(temp_dataTemp.tostring())
   
        f.close()
print("Run Over")


    
    
```
