```python
def plot_scatter():
    plt.figure(figsize=(10,8),dpi=256)
    colors = ['red', 'blue', 'green', 'gray', 'yellow', 'cyan', 'pink']
    region = ['工业园区','吴中','姑苏','高新', '相城', '吴江', '昆山']
    markers = ['o','s','v','x','^','p','d']
#     print(district)
    for i in range(7):
        x = house.loc[house['region'] == region[i]]['area']
        y = house.loc[house['region'] == region[i]]['avg_price']
        plt.scatter(x, y, c=colors[i], s=20, label=region[i], marker=markers[i])
 
 
    plt.legend(loc=1,bbox_to_anchor=(1.138,1.0),fontsize=12)
    plt.xlim(0,500)
    plt.ylim(0,150000)
    plt.title('苏州各行政区内房屋面积对房价的影响（散点图）',fontsize=20)
    plt.xlabel('房屋面积（平方米）',fontsize=16)
    plt.ylabel('房屋单价（元/平方米）',fontsize=16)
    plt.show()
 
plot_scatter()
```

```python
#导入优化模块，作线性拟合
from scipy import optimize
#直线函数方程
def linearfitting(x, A, B):
    return B + (A * x)
 
def plot_line():
    plt.figure(figsize=(10,8),dpi=256)
    colors = ['red', 'blue', 'green', 'gray', 'yellow', 'cyan', 'pink']
    region = ['工业园区','吴中','姑苏','高新', '相城', '吴江', '昆山']
    markers = ['o','s','v','x','^','p','d']

    for i in range(7):
        x = house.loc[house['region'] == region[i]]['area']
        y = house.loc[house['region'] == region[i]]['avg_price']
        A, B = optimize.curve_fit(linearfitting, x, y)[0]
        xx = np.arange(0, 2000, 100)
        yy = A * xx + B
        plt.plot(xx, yy, c=colors[i], marker=markers[i],label=region[i],linewidth=2)
 
 
    plt.legend(loc=1,bbox_to_anchor=(1.138,1.0),fontsize=12)
    plt.xlim(0,2000)
    plt.ylim(0,100000)
    plt.title('苏州各行政区内房屋面积对房价的影响（线性拟合）',fontsize=20)
    plt.xlabel('房屋面积（平方米）',fontsize=16)
    plt.ylabel('房屋单价（元/平方米）',fontsize=16)
    plt.show()
 
plot_line()
```

```python
from sklearn import linear_model
region = ['工业园区','吴中','姑苏','高新', '相城', '吴江', '昆山']
x = house.loc[house['region'] == region[0]]['area']
y = house.loc[house['region'] == region[0]]['avg_price']
length =len(x)
datasets_X= np.array(x).reshape([length,1])
#将datasets_Y转化为数组
datasets_Y=np.array(y)
```

```python
minX =min(datasets_X)
maxX =max(datasets_X)
#以数据datasets_X的最大值和最小值为范围，建立等差数列，方便后续画图。
X=np.arange(minX,maxX).reshape([-1,1])
```

```python
linear =linear_model.LinearRegression()
linear.fit(datasets_X,datasets_Y)#调用线性回归模块，建立回归方程，拟合数据
#查看回归方程系数
print('Cofficients:',linear.coef_)
#查看回归方程截距
print('intercept',linear.intercept_)
```

```python
#3.可视化处理
#scatter函数用于绘制数据 点，这里表示用红色绘制数据点；
plt.scatter(datasets_X,datasets_Y,color='red')
#plot函数用来绘制直线，这 里表示用蓝色绘制回归线；
#xlabel和ylabel用来指定横纵坐标的名称
plt.plot(X,linear.predict(X),color='blue')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
import requests
import json
ak = "ILz5CGyVIjNhpmG8NgRA68LMQRkDpRlP"
gov_locations = ['苏州工业园区管理委员会', '吴中区政府', '姑苏区政府', '苏州高新区管理委员会', '相城区政府', '吴江区政府', '昆山市人民政府', '苏州市政府']
gov_info = []
base_url = "http://api.map.baidu.com/geocoding/v3/?address="
end_url = "&city=苏州市&output=json&ak=" + str(ak) + "&callback=showLocation"
for it in gov_locations:
    url = base_url + it + end_url
    res = requests.get(url=url)
    gov_info.append(json.loads(res.text[27:-1]))
```

```python
gov_info
```

```python
gov_posi = [[it['result']['location']['lng'],it['result']['location']['lat']] for it in gov_info]
gov_posi
```

```python
def plot_scatter_gov():
    plt.figure(figsize=(10,8),dpi=256)
    colors = ['red', 'blue', 'green', 'gray', 'yellow', 'cyan', 'pink']
    region = ['工业园区','吴中','姑苏','高新', '相城', '吴江', '昆山']
    markers = ['o','s','v','x','^','p','d']
#     print(district)
    for i in range(7):
        p = np.array(house.loc[house['region'] == region[i]]['position'].str.split(','))
        x = [min((geodesic((p[i][1],p[i][0]), (gov_posi[j][1],gov_posi[j][0])).km) for j in range(len(gov_posi))) for i in range(len(p))]
        y = np.array(house.loc[house['region'] == region[i]]['avg_price'])
        plt.scatter(x, y, c=colors[i], s=20, label=region[i], marker=markers[i])
 
 
    plt.legend(loc=1,bbox_to_anchor=(1.138,1.0),fontsize=12)
    plt.xlim(0,25)
    plt.ylim(0,100000)
    plt.title('苏州与最近政府的距离对房价的影响（散点图）',fontsize=20)
    plt.xlabel('房屋面积（平方米）',fontsize=16)
    plt.ylabel('房屋单价（元/平方米）',fontsize=16)
    plt.show()
 
plot_scatter_gov()
```
